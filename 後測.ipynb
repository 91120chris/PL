```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('adult.csv')

income = df['Income'].values

mean_income = np.mean(income)
std_income = np.std(income)
median_income = np.median(income)

print("平均值:", mean_income)
print("標準差:", std_income)
print("中位數:", median_income)

plt.hist(income, bins=20, color='skyblue', edgecolor='black')
plt.axvline(mean_income, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median_income, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
