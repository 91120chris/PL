```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('2021-2022 NBA Player Stats - Playoffs.csv')
df.head()
```

```python
df2 = pd.read_csv('2021-2022 NBA Warriors Player Stats.csv')
df2.head()
```

```python
pl=set(df.iloc[:,1])
print(pl)
wa=set(df2.iloc[:,1])
print(wa)
```

```python
union_set = pl.union(wa)
intersection_set = pl.intersection(wa)
difference_set_a_b = pl.difference(wa)

print("聯集:", union_set)
print("交集:", intersection_set)
print("差集 pl-wa:", difference_set_a_b)
```

分析:

聯集結果為: 全部有打NBA季後賽的球員

交集結果為: 有打NBA季後賽的勇士隊球員

差集結果為: 有打NBA季後賽的非勇士隊球員
